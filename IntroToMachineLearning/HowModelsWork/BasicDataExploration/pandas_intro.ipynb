```python
import pandas as pd

melbourne_file_path = '../input/melbourne-housing-snapshot/melb_data.csv'

# 変数/列を選択するには、データセット内のすべての列のリストを確認する必要があります。
melbourne_data = pd.read_csv(melbourne_file_path)
melbourne_data.columns
```

```python
# dropna は欠損値を削除します（na は「利用不可」を意味します）。
melbourne_data = melbourne_data.dropna(axis=0)
print(melbourne_data)
```

```python
# ドット記法を使って予測したい列を選択します。
# これを予測対象と呼びます。
# 慣例的に、予測対象はyと呼ばれます。

y = melbourne_data.Price
print(y)
```

```python
# 複数の特徴量を選択するには、括弧で囲んだ列名のリストを指定します。

melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
print(melbourne_features)
```

```python
# 慣例により、このデータは X と呼ばれます。

x = melbourne_data[melbourne_features]
print(x)
```

```python
# 住宅価格を予測するために使用するデータを表示
x.describe()
```

```python
# 上位数行を表示する
x.head()
```

```python
# scikit-learn は、通常 DataFrame に格納される種類のデータをモデリングするための最も人気のあるライブラリです。

from sklearn.tree import DecisionTreeRegressor

melbourne_model = DecisionTreeRegressor(random_state=1)

y = melbourne_data.Price

X = melbourne_data[melbourne_features]
melbourne_model.fit(x, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```
